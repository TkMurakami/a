```python
startYear = 2000
endYear = 2020
start = 2000
end = 2020

# 取得期間を指定（例：2000年1月1日から2020年12月31日）
start_date = datetime(2000, 1, 1)
end_date = datetime(2020, 12, 31)

# Load Landsat image collections
L5coll = ee.ImageCollection('LANDSAT/LT05/C01/T1_SR') \
    .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa']) \
    .filterBounds(geometry)\
    .filterDate(start_date, end_date)


L7coll = ee.ImageCollection('LANDSAT/LE07/C01/T1_SR') \
    .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa']) \
    .filterBounds(geometry)\
    .filterDate(start_date, end_date)

L8coll = ee.ImageCollection('LANDSAT/LC08/C01/T1_SR') \
    .filterBounds(geometry) \
    .map(lambda image: image.rename(['B0', 'B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'B6', 'B8', 'B9', 'pixel_qa', 'B11'])) \
    .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa'])\
    .filterDate(start_date, end_date)

collection_merge = ee.ImageCollection(L5coll.merge(L7coll.merge(L8coll)))

# Function to calculate NDVI
def calc_ndvi(image):
    ndvi = image.normalizedDifference(['B4', 'B3']).rename('NDVI')
    return image.addBands(ndvi.copyProperties(image, ['system:time_start']))

# Function to calculate EVI
def calc_evi(image):
    B1 = image.select('B1')
    B3 = image.select('B3')
    B4 = image.select('B4')
    G = 2.5  # Gain factor
    C1 = 6.0  # Coefficient 1
    C2 = 7.5  # Coefficient 2
    L = 1.0  # Canopy background adjustment
    denominator = B4.add(B3.multiply(C1)).subtract(B1.multiply(C2)).add(L)
    evi = B4.subtract(B3).multiply(G).divide(denominator).rename('EVI')
    return image.addBands(evi.copyProperties(image, ['system:time_start']))

# Function to calculate NBR
def calc_nbr(image):
    B4 = image.select('B4')
    B7 = image.select('B7')
    nbr = (B4.subtract(B7)).divide(B4.add(B7)).rename('NBR')
    return image.addBands(nbr.copyProperties(image, ['system:time_start']))


    # Function to calculate NDBI
def calc_ndbi(image):
    B4 = image.select('B4')
    B5 = image.select('B5')
    ndbi = B5.subtract(B4).divide(B5.add(B4)).rename('NDBI')
    return image.addBands(ndbi.copyProperties(image, ['system:time_start']))

# Cloud masking functions
def cloud_mask(image):
    qa_mask = image.select('pixel_qa')
    qa_mask_0 = qa_mask.bitwiseAnd(1 << 5).eq(0)
    return image.updateMask(qa_mask_0)

def cloud_shadow_mask(image):
    qa_mask = image.select('pixel_qa')
    qa_mask_0 = qa_mask.bitwiseAnd(1 << 3).eq(0)
    return image.updateMask(qa_mask_0)

def snow_mask(image):
    qa_mask = image.select('pixel_qa')
    qa_mask_0 = qa_mask.bitwiseAnd(1 << 4).eq(0)
    return image.updateMask(qa_mask_0)

# Apply cloud masking and band calculations
collection_merge = (collection_merge
                    .map(cloud_mask)
                    .map(cloud_shadow_mask)
                    .map(snow_mask)
                    .map(calc_ndvi)
                    .map(calc_evi)
                    .map(calc_nbr)
                    .map(calc_ndbi))

# Define parameters
min_observations = [2,4,6,8]
chi_square_probability = [0.8, 0.9, 0.99]
min_num_of_years_scaler = [1.00, 1.33, 2.00]
lambda_value = [0, 20]
max_iterations = [500]
bands = [['B1'], ['B2'], ['B3']]






# Create an empty feature collection to store results
output_collection1 = ee.FeatureCollection([])

# Generate combinations of parameters
parameter_combinations = [(obs, prob, scaler, l, max_iter,band)
                           for band in bands
                           for obs in min_observations
                           for prob in chi_square_probability
                           for scaler in min_num_of_years_scaler
                           for l in lambda_value
                           for max_iter in max_iterations
                          ]

# Temporary list to store results
results_list = []

# Loop over parameter combinations
for params in parameter_combinations:
    min_obs, chi_square_prob, min_years_scaler, lambda_val, max_iter, band = params

    # Run CCD
    ccd_results = ee.Algorithms.TemporalSegmentation.Ccdc(
        breakpointBands = band,
        # tmaskBands = ['B2', 'B7'],
        minObservations = min_obs,
        chiSquareProbability = chi_square_prob,
        minNumOfYearsScaler = min_years_scaler,
        dateFormat = 1,
        # lambda = lambda_val,
        maxIterations = max_iter,
        collection = collection_merge
    )
    t_break = ccd_results.select('tBreak').toFloat().toInt()
    change_prob = ccd_results.select('changeProb')
    mask = t_break.gt(start).And(t_break.lte(end)).And(change_prob.eq(1))

    # Get change years from CCD results
    change_years_masked = t_break.arrayMask(mask)

    # Convert to numeric array
    features = ee.FeatureCollection(
        ee.Array(change_years_masked
                 .reduceRegion(reducer=ee.Reducer.first(),
                               geometry=geometry,
                               scale=30)
                 .get('tBreak'))
        .toList()
        .map(lambda row: ee.Feature(geometry, {'year': row})))

    # Filter results based on change prediction for the change year
    feature_0 = features.filter(
        ee.Filter.Or(
              ee.Filter.eq('year', 2000)
        ))
    feature_1 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2001)
        ))
    feature_2 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2002)
        ))
    feature_3 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2003)
        ))
    feature_4 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2004)
        ))
    feature_5 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2005)
        ))
    feature_6 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2006)
        ))
    feature_7 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2007)
        ))
    feature_8 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2008)
        ))
    feature_9 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2009)
        ))
    feature_10 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2010)
        ))
    feature_11 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2011)
        ))
    feature_12 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2012)
        ))
    feature_13 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2013)
        ))
    feature_14 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2014)
        ))
    feature_15 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2015)
        ))
    feature_16 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2016)
        ))
    feature_17 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2017)
        ))
    feature_18 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2018)
        ))
    feature_19 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2019)
        ))
    feature_20 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2020)
        ))

    feature_0_20 = features.filter(
        ee.Filter.Or(
            ee.Filter.eq('year', 2000),
            ee.Filter.eq('year', 2001),
            ee.Filter.eq('year', 2002),
            ee.Filter.eq('year', 2003),
            ee.Filter.eq('year', 2004),
            ee.Filter.eq('year', 2005),
            ee.Filter.eq('year', 2006),
            ee.Filter.eq('year', 2007),
            ee.Filter.eq('year', 2008),
            ee.Filter.eq('year', 2009),
            ee.Filter.eq('year', 2010),
            ee.Filter.eq('year', 2011),
            ee.Filter.eq('year', 2012),
            ee.Filter.eq('year', 2013),
            ee.Filter.eq('year', 2014),
            ee.Filter.eq('year', 2015),
            ee.Filter.eq('year', 2016),
            ee.Filter.eq('year', 2017),
            ee.Filter.eq('year', 2018),
            ee.Filter.eq('year', 2019),
            ee.Filter.eq('year', 2020)
        ))


    # Determine if the prediction matches the actual change year
    feature_0  = feature_0.size().eq(0)
    feature_1  = feature_1.size().eq(0)
    feature_2  = feature_2.size().eq(0)
    feature_3  = feature_3.size().eq(0)
    feature_4  = feature_4.size().eq(0)
    feature_5  = feature_5.size().eq(0)
    feature_6  = feature_6.size().eq(0)
    feature_7  = feature_7.size().eq(0)
    feature_8  = feature_8.size().eq(0)
    feature_9  = feature_9.size().eq(0)
    feature_10  = feature_10.size().eq(0)
    feature_11  = feature_11.size().eq(0)
    feature_12  = feature_12.size().eq(0)
    feature_13  = feature_13.size().eq(0)
    feature_14  = feature_14.size().eq(0)
    feature_15  = feature_15.size().eq(0)
    feature_16  = feature_16.size().eq(0)
    feature_17  = feature_17.size().eq(0)
    feature_18  = feature_18.size().eq(0)
    feature_19  = feature_19.size().eq(0)
    feature_20  = feature_20.size().eq(0)
    feature_0_20 = feature_0_20.size().eq(0)


    # Create feature for the current parameter combination
    feature = ee.Feature(None, {
        'band': band,
        'minObs': min_obs,
        'chiSquareProb': chi_square_prob,
        'minYearsScaler': min_years_scaler,
        'lambdaValue': lambda_val,
        'maxIter': max_iter,
        'feature_0_20':feature_0_20,
        'feature_0':feature_0,
        'feature_1':feature_1,
        'feature_2':feature_2,
        'feature_3':feature_3,
        'feature_4':feature_4,
        'feature_5':feature_5,
        'feature_6':feature_6,
        'feature_7':feature_7,
        'feature_8':feature_8,
        'feature_9':feature_9,
        'feature_10':feature_10,
        'feature_11':feature_11,
        'feature_12':feature_12,
        'feature_13':feature_13,
        'feature_14':feature_14,
        'feature_15':feature_15,
        'feature_16':feature_16,
        'feature_17':feature_17,
        'feature_18':feature_18,
        'feature_19':feature_19,
        'feature_20':feature_20

    })

    # Add feature to the output collection
    output_collection1 = output_collection1.merge(ee.FeatureCollection([feature]))


```
