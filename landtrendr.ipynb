```python


startYear = 2000
endYear = 2020

changeYear = changeYear -1

preChangeYear = changeYear - 1
nextChangeYear = changeYear + 1

preChangeYear_2 = changeYear - 2
nextChangeYear_2 = changeYear + 2

preChangeYear_3 = changeYear - 3
preChangeYear_4 = changeYear - 4

L5coll = (ee.ImageCollection('LANDSAT/LT05/C01/T1_SR')
          .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa'])
          .filterBounds(geometry))

L7coll = (ee.ImageCollection('LANDSAT/LE07/C01/T1_SR')
          .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa'])
          .filterBounds(geometry))

L8coll = (ee.ImageCollection('LANDSAT/LC08/C01/T1_SR')
          .filterBounds(geometry)
          .map(lambda image: image.rename(['B0', 'B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'B6', 'B8', 'B9', 'pixel_qa', 'B11']))
          .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa']))

collection_merge = L5coll.merge(L7coll.merge(L8coll))


def calcNdvi(image):
    B3 = image.select('B4')
    B4 = image.select('B3')
    ndvi = (B4.subtract(B3)).divide(B4.add(B3)).rename('NDVI')
    return image.addBands(ndvi.copyProperties(image, ['system:time_start']))

def calcEvi(image):
    B1 = image.select('B1')
    B3 = image.select('B3')
    B4 = image.select('B4')
    G = 2.5
    C1 = 6.0
    C2 = 7.5
    L = 1.0
    denominator = (B4.add(B3.multiply(C1))
                   .subtract(B1.multiply(C2))
                   .add(L))
    evi = (B4.subtract(B3).multiply(G)).divide(denominator).rename('EVI')
    return image.addBands(evi.copyProperties(image, ['system:time_start']))

def calcNbr(image):
    B4 = image.select('B4')
    B7 = image.select('B7')
    nbr = (B4.subtract(B7)).divide(B4.add(B7)).rename('NBR')
    return image.addBands(nbr.copyProperties(image, ['system:time_start']))



    # Function to calculate NDBI
def calc_ndbi(image):
    B4 = image.select('B4')
    B5 = image.select('B5')
    ndbi = B5.subtract(B4).divide(B5.add(B4)).rename('NDBI')
    return image.addBands(ndbi.copyProperties(image, ['system:time_start']))

def cloudMask(image):
    qaMask = image.select('pixel_qa')
    qamask0 = qaMask.bitwiseAnd(1 << 5).eq(0)
    return image.updateMask(qamask0)

def cloudShadowMask(image):
    qaMask = image.select('pixel_qa')
    qamask0 = qaMask.bitwiseAnd(1 << 3).eq(0)
    return image.updateMask(qamask0)

def snowMask(image):
    qaMask = image.select('pixel_qa')
    qamask0 = qaMask.bitwiseAnd(1 << 4).eq(0)
    return image.updateMask(qamask0)

collection_merge = (collection_merge
                    .map(cloudMask)
                    .map(cloudShadowMask)
                    .map(snowMask)
                    .map(calcNdvi)
                    .map(calcEvi)
                    .map(calcNbr)
                    .map(calc_ndbi))


def compositeByYear(year):
    startDate = ee.Date.fromYMD(year, 1, 1)
    endDate = startDate.advance(1, 'year')

    return (collection_merge
            .filterDate(startDate, endDate)
            .median()
            .set('year', year)
            .set('system:time_start', startDate.millis()))


imageList = ee.List.sequence(startYear, endYear).map(compositeByYear)

yearlyComposites = ee.ImageCollection.fromImages(imageList)

# Define bands and parameters

bands1 = ['B1', 'B2', 'B3',]
bands2 = ['B4', 'B5', 'B7']
bands3 = ['NDVI', 'EVI', 'NBR', 'NDBI']

MaxSegments = [3.0, 4.0, 5.0]
PvalThreshold = [0.05, 0.10,0.15]
SpikeThreshold = [0.75, 0.85, 0.95]
RecoveryThreshold = [0.25, 0.50, 1.00]
BestModelProportion = [0.75, 0.85, 0.95]

featureCollection1 = ee.FeatureCollection([])

for bandName in bands1:
    yearlyComposites_band = yearlyComposites.select(bandName)

    for n_MaxSegment in MaxSegments:
        for n_SpikeThreshold in SpikeThreshold:
            for n_RecoveryThreshold in RecoveryThreshold:
                for n_PvalThreshold in PvalThreshold:
                    for n_BestModelProportion in BestModelProportion:

                        lt = ee.Algorithms.TemporalSegmentation.LandTrendr(
                            timeSeries=yearlyComposites_band,
                            maxSegments=n_MaxSegment,
                            spikeThreshold=n_SpikeThreshold,
                            vertexCountOvershoot=3,
                            preventOneYearRecovery=True,
                            recoveryThreshold=n_RecoveryThreshold,
                            pvalThreshold=n_PvalThreshold,
                            bestModelProportion=n_BestModelProportion,
                            minObservationsNeeded=6
                        ).clip(geometry)

                        ltarray = lt.select(['LandTrendr'])


                        def createFeature(row):
                            row = ee.List(row)
                            year = ee.Number(row.get(0))  # 'year'に対応する値を数値として取得
                            vertex = ee.Number(row.get(3))  # 'vertex'に対応する値を数値として取得
                            return ee.Feature(geometry, {
                                'year': year,
                                'vertex': vertex,
                            })

                        features = ee.FeatureCollection(ee.Array(ltarray.reduceRegion(
                                        reducer=ee.Reducer.first(),
                                        geometry=geometry,
                                        scale=30)
                                    .get('LandTrendr')).transpose().toList().map(createFeature)
                                    )
                        # print(features.getInfo())



                        filteredResults_buf0 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', changeYear),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )

                        filteredResults_buf1 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear),
                                    ee.Filter.eq('year', changeYear),
                                    ee.Filter.eq('year', preChangeYear_2)
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )



                        filteredResults_pre1 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_pre2 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear_2),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_pre3 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear_3),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_pre4 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear_4),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_next1 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', nextChangeYear)
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_next2 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', nextChangeYear_2)
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )


                        filteredResults_buf0 = filteredResults_buf0.size().gte(1)
                        filteredResults_buf1 = filteredResults_buf1.size().gte(1)
                        filteredResults_pre1 = filteredResults_pre1.size().gte(1)
                        filteredResults_pre2 = filteredResults_pre2.size().gte(1)
                        filteredResults_pre3 = filteredResults_pre3.size().gte(1)
                        filteredResults_pre4 = filteredResults_pre4.size().gte(1)
                        filteredResults_next1 = filteredResults_next1.size().gte(1)
                        filteredResults_next2 = filteredResults_next2.size().gte(1)
                        new_feature = ee.Feature(geometry, {
                            'Band': bandName,
                            'MaxSegments': n_MaxSegment,
                            'PvalThreshold': n_PvalThreshold,
                            'SpikeThreshold': n_SpikeThreshold,
                            'RecoveryThreshold': n_RecoveryThreshold,
                            'BestModelProportion': n_BestModelProportion,
                            'filteredResults_buf0':filteredResults_buf0,
                            'filteredResults_buf1':filteredResults_buf1,
                            'filteredResults_pre1':filteredResults_pre1,
                            'filteredResults_pre2':filteredResults_pre2,
                            'filteredResults_pre3':filteredResults_pre3,
                            'filteredResults_pre4':filteredResults_pre4,
                            'filteredResults_next1':filteredResults_next1,
                            'filteredResults_next2':filteredResults_next2
                        })

                        featureCollection1 = featureCollection1.merge(new_feature)


```

```python
featureCollection3 = ee.FeatureCollection([])

for bandName in bands3:
    yearlyComposites_band = yearlyComposites.select(bandName)

    for n_MaxSegment in MaxSegments:
        for n_SpikeThreshold in SpikeThreshold:
            for n_RecoveryThreshold in RecoveryThreshold:
                for n_PvalThreshold in PvalThreshold:
                    for n_BestModelProportion in BestModelProportion:

                        lt = ee.Algorithms.TemporalSegmentation.LandTrendr(
                            timeSeries=yearlyComposites_band,
                            maxSegments=n_MaxSegment,
                            spikeThreshold=n_SpikeThreshold,
                            vertexCountOvershoot=3,
                            preventOneYearRecovery=True,
                            recoveryThreshold=n_RecoveryThreshold,
                            pvalThreshold=n_PvalThreshold,
                            bestModelProportion=n_BestModelProportion,
                            minObservationsNeeded=6
                        ).clip(geometry)

                        ltarray = lt.select(['LandTrendr'])


                        def createFeature(row):
                            row = ee.List(row)
                            year = ee.Number(row.get(0))  # 'year'に対応する値を数値として取得
                            vertex = ee.Number(row.get(3))  # 'vertex'に対応する値を数値として取得
                            return ee.Feature(geometry, {
                                'year': year,
                                'vertex': vertex,
                            })

                        features = ee.FeatureCollection(ee.Array(ltarray.reduceRegion(
                                        reducer=ee.Reducer.first(),
                                        geometry=geometry,
                                        scale=30)
                                    .get('LandTrendr')).transpose().toList().map(createFeature)
                                    )

                        filteredResults_buf0 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', changeYear),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )

                        filteredResults_buf1 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear),
                                    ee.Filter.eq('year', changeYear),
                                    ee.Filter.eq('year', preChangeYear_2)
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )



                        filteredResults_pre1 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_pre2 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear_2),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_pre3 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear_3),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_pre4 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', preChangeYear_4),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_next1 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', nextChangeYear)
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        filteredResults_next2 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', nextChangeYear_2)
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )


                        filteredResults_buf0 = filteredResults_buf0.size().gte(1)
                        filteredResults_buf1 = filteredResults_buf1.size().gte(1)
                        filteredResults_pre1 = filteredResults_pre1.size().gte(1)
                        filteredResults_pre2 = filteredResults_pre2.size().gte(1)
                        filteredResults_pre3 = filteredResults_pre3.size().gte(1)
                        filteredResults_pre4 = filteredResults_pre4.size().gte(1)
                        filteredResults_next1 = filteredResults_next1.size().gte(1)
                        filteredResults_next2 = filteredResults_next2.size().gte(1)
                        new_feature = ee.Feature(geometry, {
                            'Band': bandName,
                            'MaxSegments': n_MaxSegment,
                            'PvalThreshold': n_PvalThreshold,
                            'SpikeThreshold': n_SpikeThreshold,
                            'RecoveryThreshold': n_RecoveryThreshold,
                            'BestModelProportion': n_BestModelProportion,
                            'filteredResults_buf0':filteredResults_buf0,
                            'filteredResults_buf1':filteredResults_buf1,
                            'filteredResults_pre1':filteredResults_pre1,
                            'filteredResults_pre2':filteredResults_pre2,
                            'filteredResults_pre3':filteredResults_pre3,
                            'filteredResults_pre4':filteredResults_pre4,
                            'filteredResults_next1':filteredResults_next1,
                            'filteredResults_next2':filteredResults_next2
                        })

                        featureCollection3 = featureCollection3.merge(new_feature)
```
