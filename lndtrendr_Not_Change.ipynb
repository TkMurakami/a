```python


startYear = 2000
endYear = 2020

startDate = ee.Date.fromYMD(startYear, 1, 1)
endDate =  ee.Date.fromYMD(endYear, 12, 31)


#landsatのデータ取得
L5coll = (ee.ImageCollection('LANDSAT/LT05/C01/T1_SR')
          .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa'])
          .filterDate(startDate, endDate)
          .filterBounds(geometry))

L7coll = (ee.ImageCollection('LANDSAT/LE07/C01/T1_SR')
          .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa'])
          .filterDate(startDate, endDate)
          .filterBounds(geometry))

L8coll = (ee.ImageCollection('LANDSAT/LC08/C01/T1_SR')
          .filterBounds(geometry)
          .map(lambda image: image.rename(['B0', 'B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'B6', 'B8', 'B9', 'pixel_qa', 'B11']))
          .filterDate(startDate, endDate)
          .select(['B1', 'B2', 'B3', 'B4', 'B5', 'B7', 'pixel_qa']))

collection_merge = L5coll.merge(L7coll.merge(L8coll))



#インデックス計算用関数
def calcNdvi(image):
    B3 = image.select('B4')
    B4 = image.select('B3')
    ndvi = (B4.subtract(B3)).divide(B4.add(B3)).rename('NDVI')
    return image.addBands(ndvi.copyProperties(image, ['system:time_start']))

def calcEvi(image):
    B1 = image.select('B1')
    B3 = image.select('B3')
    B4 = image.select('B4')
    G = 2.5
    C1 = 6.0
    C2 = 7.5
    L = 1.0
    denominator = (B4.add(B3.multiply(C1))
                   .subtract(B1.multiply(C2))
                   .add(L))
    evi = (B4.subtract(B3).multiply(G)).divide(denominator).rename('EVI')
    return image.addBands(evi.copyProperties(image, ['system:time_start']))

def calcNbr(image):
    B4 = image.select('B4')
    B7 = image.select('B7')
    nbr = (B4.subtract(B7)).divide(B4.add(B7)).rename('NBR')
    return image.addBands(nbr.copyProperties(image, ['system:time_start']))

def calcNdmi(image):
    B4 = image.select('B4')
    B5 = image.select('B5')
    ndmi = (B4.subtract(B5)).divide(B4.add(B5)).rename('NDMI')
    return image.addBands(ndmi.copyProperties(image, ['system:time_start']))

    # Function to calculate NDBI
def calc_ndbi(image):
    B4 = image.select('B4')
    B5 = image.select('B5')
    ndbi = B5.subtract(B4).divide(B5.add(B4)).rename('NDBI')
    return image.addBands(ndbi.copyProperties(image, ['system:time_start']))

def cloudMask(image):
    qaMask = image.select('pixel_qa')
    qamask0 = qaMask.bitwiseAnd(1 << 5).eq(0)
    return image.updateMask(qamask0)

def cloudShadowMask(image):
    qaMask = image.select('pixel_qa')
    qamask0 = qaMask.bitwiseAnd(1 << 3).eq(0)
    return image.updateMask(qamask0)

def snowMask(image):
    qaMask = image.select('pixel_qa')
    qamask0 = qaMask.bitwiseAnd(1 << 4).eq(0)
    return image.updateMask(qamask0)


#インデックス計算とノイズ除去
collection_merge = (collection_merge
                    .map(cloudMask)
                    .map(cloudShadowMask)
                    .map(snowMask)
                    .map(calcNdvi)
                    .map(calcEvi)
                    .map(calcNbr)
                    .map(calc_ndbi))


#年次時系列データにする
def compositeByYear(year):
    startDate = ee.Date.fromYMD(year, 1, 1)
    endDate =  ee.Date.fromYMD(year, 12, 31)

    return (collection_merge
            .filterDate(startDate, endDate)
            .median()
            .set('year', year)
            .set('system:time_start', startDate.millis())
    )





imageList = ee.List.sequence(startYear, endYear).map(compositeByYear)

yearlyComposites = ee.ImageCollection.fromImages(imageList)

# Define bands and parameters
bands1 = ['B1', 'B2', 'B3',]
bands2 = ['B4', 'B5', 'B7']
bands3 = ['NDVI', 'EVI']
bands4 = ['NBR', 'NDBI']
MaxSegments = [3.0, 4.0, 5.0]
PvalThreshold = [0.05, 0.10,0.15]
SpikeThreshold = [0.75, 0.85, 0.95]
RecoveryThreshold = [0.25, 0.50, 1.00]
BestModelProportion = [0.75, 0.85, 0.95]


featureCollection1 = ee.FeatureCollection([])

for bandName in bands1:
    yearlyComposites_band = yearlyComposites.select(bandName)

    for n_MaxSegment in MaxSegments:
        for n_SpikeThreshold in SpikeThreshold:
            for n_RecoveryThreshold in RecoveryThreshold:
                for n_PvalThreshold in PvalThreshold:
                    for n_BestModelProportion in BestModelProportion:

                        lt = ee.Algorithms.TemporalSegmentation.LandTrendr(
                            timeSeries=yearlyComposites_band,
                            maxSegments=n_MaxSegment,
                            spikeThreshold=n_SpikeThreshold,
                            vertexCountOvershoot=3,
                            preventOneYearRecovery=True,
                            recoveryThreshold=n_RecoveryThreshold,
                            pvalThreshold=n_PvalThreshold,
                            bestModelProportion=n_BestModelProportion,
                            minObservationsNeeded=6
                        ).clip(geometry)

                        ltarray = lt.select(['LandTrendr'])


                        def createFeature(row):
                            row = ee.List(row)
                            year = ee.Number(row.get(0))  # 'year'に対応する値を数値として取得
                            vertex = ee.Number(row.get(3))  # 'vertex'に対応する値を数値として取得
                            return ee.Feature(geometry, {
                                'year': year,
                                'vertex': vertex,
                            })

                        features = ee.FeatureCollection(ee.Array(ltarray.reduceRegion(
                                        reducer=ee.Reducer.first(),
                                        geometry=geometry,
                                        scale=30)
                                    .get('LandTrendr')).transpose().toList().map(createFeature)
                                    )


                        feature_1999 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 1999),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )

                        feature_2000 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2000),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2001 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2001),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2002 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2002),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2003 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2003),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2004 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2004),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2005 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2005),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2006 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2006),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2007 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2007),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2008 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2008),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2009 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2009),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2010 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2010),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2011 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2011),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2012 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2012),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2013 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2013),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2014 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2014),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2015 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2015),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2016 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2016),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2017 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2017),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2018 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2018),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2019 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2019),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2020 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2020),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )
                        feature_2021 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2021),
                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )

                        feature_0_20 = features.filter(
                            ee.Filter.And(
                                ee.Filter.Or(
                                    ee.Filter.eq('year', 2000),
                                    ee.Filter.eq('year', 2001),
                                    ee.Filter.eq('year', 2002),
                                    ee.Filter.eq('year', 2003),
                                    ee.Filter.eq('year', 2004),
                                    ee.Filter.eq('year', 2005),
                                    ee.Filter.eq('year', 2006),
                                    ee.Filter.eq('year', 2007),
                                    ee.Filter.eq('year', 2008),
                                    ee.Filter.eq('year', 2009),
                                    ee.Filter.eq('year', 2010),
                                    ee.Filter.eq('year', 2011),
                                    ee.Filter.eq('year', 2012),
                                    ee.Filter.eq('year', 2013),
                                    ee.Filter.eq('year', 2014),
                                    ee.Filter.eq('year', 2015),
                                    ee.Filter.eq('year', 2016),
                                    ee.Filter.eq('year', 2017),
                                    ee.Filter.eq('year', 2018)

                                ),
                                ee.Filter.eq('vertex', 1)
                            )
                        )


                        feature_1999 = feature_1999.size().eq(0)
                        feature_2000 = feature_2000.size().eq(0)
                        feature_2001 = feature_2001.size().eq(0)
                        feature_2002 = feature_2002.size().eq(0)
                        feature_2003 = feature_2003.size().eq(0)
                        feature_2004 = feature_2004.size().eq(0)
                        feature_2005 = feature_2005.size().eq(0)
                        feature_2006 = feature_2006.size().eq(0)
                        feature_2007 = feature_2007.size().eq(0)
                        feature_2008= feature_2008.size().eq(0)
                        feature_2009 = feature_2009.size().eq(0)
                        feature_2010 = feature_2010.size().eq(0)
                        feature_2011 = feature_2011.size().eq(0)
                        feature_2012 = feature_2012.size().eq(0)
                        feature_2013 = feature_2013.size().eq(0)
                        feature_2014 = feature_2014.size().eq(0)
                        feature_2015 = feature_2015.size().eq(0)
                        feature_2016 = feature_2016.size().eq(0)
                        feature_2017 = feature_2017.size().eq(0)
                        feature_2018 = feature_2018.size().eq(0)
                        feature_2019 = feature_2019.size().eq(0)
                        feature_2020 = feature_2020.size().eq(0)
                        feature_2021 = feature_2021.size().eq(0)
                        feature_0_20 = feature_0_20.size().eq(0)
                        new_feature = ee.Feature(geometry, {
                            'Band': bandName,
                            'MaxSegments': n_MaxSegment,
                            'PvalThreshold': n_PvalThreshold,
                            'SpikeThreshold': n_SpikeThreshold,
                            'RecoveryThreshold': n_RecoveryThreshold,
                            'BestModelProportion': n_BestModelProportion,
                            'feature_0_20':feature_0_20,
                            'feature_1999':feature_1999,
                            'feature_2000':feature_2000,
                            'feature_2001':feature_2001,
                            'feature_2002':feature_2002,
                            'feature_2003':feature_2003,
                            'feature_2004':feature_2004,
                            'feature_2005':feature_2005,
                            'feature_2006':feature_2006,
                            'feature_2007':feature_2007,
                            'feature_2008':feature_2008,
                            'feature_2009':feature_2009,
                            'feature_2010':feature_2010,
                            'feature_2011':feature_2011,
                            'feature_2012':feature_2012,
                            'feature_2013':feature_2013,
                            'feature_2014':feature_2014,
                            'feature_2015':feature_2015,
                            'feature_2016':feature_2016,
                            'feature_2017':feature_2017,
                            'feature_2018':feature_2018,
                            'feature_2019':feature_2019,
                            'feature_2020':feature_2020,
                            'feature_2021':feature_2021,

                        })

                        featureCollection1 = featureCollection1.merge(new_feature)


```
